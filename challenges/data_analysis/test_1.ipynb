```python
import pandas as pd
import datetime as dt

df = pd.read_csv('results.csv')
display(df)
df.describe()
```

```python
df1 = df.copy()
df1['day'] = df1['date'].apply(lambda x: x[8::])
df1['mounth'] = df1['date'].apply(lambda x: x[5:7])
df1['year'] = df1['date'].apply(lambda x: x[0:4])
df1['year'] = df1['year'].astype(int)
df1 = df1.dropna()
display(df1)
```

```python
#@title (1)
 
max_mounth = max(df1['mounth'].unique())
max_mounth  
```

```python
#@title (2)
import numpy as np

df1['label'] =1
table = pd.pivot_table(df1, values=['home_score','away_score'],
                    index=['home_team', 'away_team', 'label'],
                    aggfunc=np.sum)
pd.DataFrame(table)
```

```python
df = pd.DataFrame(table.to_records())
m = df[df['label'] == max(df.label.unique())]
m['away_score'] + m['home_score']
```

```python
# (3)
from collections import Counter
# Counter(m).most_common()

selected_df = df1[(df1['year'] <= 1999) & (df1['year'] >= 1900)]
selected_df['team_win'] = selected_df.apply(lambda rec: rec['home_team'] if (rec['home_score'] >  rec['away_score']) else rec['away_team'], axis=1)
winner_teams = list(selected_df['team_win'])
Counter(winner_teams).most_common()
```

```python
# (4)

df4 = selected_df.copy()
away_teams = list(selected_df['away_team'])
chosed_at = Counter(away_teams).most_common()
teams_list = []
for item in chosed_at:
    if item[1] > 50:
        teams_list += [item[0]]

df4 = df4[df4['away_team'].isin(teams_list)] 
df4['away_winner'] = selected_df.apply(lambda rec: 0 if (rec['home_score'] >  rec['away_score']) else rec['away_team'], axis=1)
win_out_team = Counter(df4['away_winner'].unique()).most_common()
win_out_team
```

```python
# (5)

df5 = df1.copy()
df5 = df5[(df5['home_team'] == 'Iran') | ( df5['away_team'] == 'Iran') ]
# sum(df5['home_team'] == 'Iran'
df5['winner'] = df5.apply(lambda rec: rec['home_team'] if (rec['home_score'] >  rec['away_score']) else rec['away_team'], axis=1)
df5 = df5[df5['winner'] == 'Iran']
df5.groupby((df5.year//10)*10).sum()
```

```python
# (6)

display(df1)
df6 = df1[df1['tournament'] != 'Friendly']
display(df6)
```

```python
df_tour = pd.DataFrame()
tournaments = df6.tournament.unique()
for item in tournaments:
    df_temp = df6[df6['tournament'] == item]
    last = df_temp.iloc[-1]
    df_tour = df_tour.append(last, ignore_index=True)

df_tour['sum_goal'] = df_tour.apply(lambda rec: rec['home_score'] + rec['away_score'], axis=1)
max(df_tour.sum_goal)
```
