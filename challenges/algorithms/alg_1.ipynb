```python
def countFrequency(string, ch):
    count = 0;
    for i in range(len(string)):
        if (string[i] == ch):
            count += 1
 
    return count
 
def frequencySort(user_input):
    n = len(user_input)
    vp = []
    for i in range(n):
        vp.append((countFrequency(user_input, user_input[i]), user_input[i]))
    vp.sort(reverse=True)

    output = []
    for i in vp:
        output += i[1]
    return "".join(output)
```

```python
from collections import OrderedDict

def checkOrder(string, pattern): 
    dic = OrderedDict.fromkeys(string) 
    ptr = 0
    sub_dic = []

    for key, value in dic.items():
        if (key == pattern[ptr]): 
            ptr = ptr + 1
            sub_dic += key
        if (ptr == (len(pattern))):
            return 1

    if ptr > 1:
        ptr = 0
        for key, value in dic.items():
            if (key == sub_dic[ptr]): 
                ptr = ptr + 1
            if (ptr == (len(sub_dic))):
                return 1
    return 0
```

```python
open_list = ["[","{","("]
close_list = ["]","}",")"]

def check(myStr):
    stack = []
    for i in myStr:
        if i in open_list:
            stack.append(i)
        elif i in close_list:
            pos = close_list.index(i)
            if ((len(stack) > 0) and
                (open_list[pos] == stack[len(stack)-1])):
                stack.pop()
            else:
                return "false"
    if len(stack) == 0:
        return "true"
    else:
        return "false"
```
